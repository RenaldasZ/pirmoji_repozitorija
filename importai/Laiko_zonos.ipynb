```python
from datetime import datetime
from pytz import timezone

# paprašome vartotojo įvesti laiką
input_time = input("Įveskite laiką HH:MM formatu: ")

# konvertuojame įvestą laiką į datetime objektą
time_obj = datetime.strptime(input_time, "%H:%M")

# paprašome vartotojo įvesti laiko zoną
timezone_input = input("Įveskite laiko zoną (pvz. Europe/Vilnius): ")

# konvertuojame laiko zoną į timezone objektą
time_zone = timezone(timezone_input)

# konvertuojame įvestą laiką į įvestos laiko zonos laiką
time_in_zone = time_zone.localize(time_obj)

# konvertuojame laiką į Sydney laiko zoną ir spausdiname
sydney_time = time_in_zone.astimezone(timezone('Australia/Sydney'))
print("Sydney laikas:", sydney_time.strftime("%H:%M"))

# konvertuojame laiką į Dubajaus laiko zoną ir spausdiname
dubai_time = time_in_zone.astimezone(timezone('Asia/Dubai'))
print("Dubajaus laikas:", dubai_time.strftime("%H:%M"))

# konvertuojame laiką į Vilniaus laiko zoną ir spausdiname
vilnius_time = time_in_zone.astimezone(timezone('Europe/Vilnius'))
print("Vilniaus laikas:", vilnius_time.strftime("%H:%M"))

# konvertuojame laiką į Londono laiko zoną ir spausdiname
london_time = time_in_zone.astimezone(timezone('Europe/London'))
print("Londono laikas:", london_time.strftime("%H:%M"))

# konvertuojame laiką į Niujorko laiko zoną ir spausdiname
new_york_time = time_in_zone.astimezone(timezone('America/New_York'))
print("Niujorko laikas:", new_york_time.strftime("%H:%M"))

# konvertuojame laiką į Los Andželo laiko zoną ir spausdiname
los_angeles_time = time_in_zone.astimezone(timezone('America/Los_Angeles'))
print("Los Andželo laikas:", los_angeles_time.strftime("%H:%M"))
```

```python
from datetime import datetime
from zoneinfo import ZoneInfo

# Paprašome vartotojo įvesti laiką
time_str = input("Įveskite laiką HH:MM formatu: ")

# Konvertuojame įvestą laiką į datetime objektą
time = datetime.strptime(time_str, "%H:%M")

# Paprašome vartotojo įvesti laiko zoną
timezone_str = input("Įveskite laiko zoną (pvz. 'Europe/Vilnius'): ")

# Konvertuojame laiko zoną į ZoneInfo objektą
timezone = ZoneInfo(timezone_str)

# Spausdiname įvestą laiką skirtingose pasaulio vietose
print("Sydney'aus laikas:", time.astimezone(ZoneInfo("Australia/Sydney")).strftime("%H:%M"))
print("Dubai'aus laikas:", time.astimezone(ZoneInfo("Asia/Dubai")).strftime("%H:%M"))
print("Vilniaus laikas:", time.astimezone(timezone).strftime("%H:%M"))
print("London'o laikas:", time.astimezone(ZoneInfo("Europe/London")).strftime("%H:%M"))
print("New York'o laikas:", time.astimezone(ZoneInfo("America/New_York")).strftime("%H:%M"))
print("Los Angeles laikas:", time.astimezone(ZoneInfo("America/Los_Angeles")).strftime("%H:%M"))
```
