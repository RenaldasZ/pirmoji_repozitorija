```python
from PIL import Image
from PIL import ImageOps

image = Image.open("background.jpeg")

cropped_img = image.crop((1000, 200, 1920, 1080))

# Apverčiame nuotrauką horizontaliai
mirrored_img = ImageOps.mirror(cropped_img)

# Rodyti apverstą nuotrauką
mirrored_img

# width, height = (600, 600)
# new_size = (width, height) # vietoje "width" ir "height" įrašykite norimus parametrus, pvz.: (60, 80)
# resized_image = image.resize(new_size)
mirrored_img.save("paveiksliukas_mirrored.jpg")


```

```python
logo = Image.open("mylogo.png").convert("RGBA")
logo.thumbnail((124, 124))
mirrored_img.paste(logo, (780, 10), logo)
mirrored_img.save("jusu_portreto_ikona.png")
mirrored_img
```
